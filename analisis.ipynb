```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data.csv")
df.tail(5)
```

```python
df["Precio"] = df["Precio"].apply(lambda x: float(x[1:].replace(',','')))
```

```python
df["Encuadernacion"].unique()
```

```python
plt.figure(figsize=(60, 60))
bp = df.boxplot(column = "Precio", by = "Genero", grid = False)
plt.title('Diagrama de Caja y Bigotes por G칠nero')
plt.xlabel('G칠nero')
plt.ylabel('Precio')
plt.xticks(rotation = 75, ha = "right")
plt.show()
```

```python
medias_por_genero = df.groupby('Genero')['Precio'].mean()
medias_por_genero
```

```python
plt.figure(figsize=(12,10))
medias_por_genero.plot(kind = "bar")
```

```python
desviaciones_por_genero = df.groupby('Genero')['Precio'].std()
desviaciones_por_genero
```

```python
desviaciones_por_genero.plot(kind = "bar")
```

```python
medias_paginas_por_genero = df.groupby('Genero')['Paginas'].mean()
medias_paginas_por_genero.plot(kind = "bar")
```

```python
plt.scatter(df["Precio"],df["Paginas"], alpha= 0.7)
```

```python
plt.figure(figsize=(60, 60))
bp = df.boxplot(column = "Paginas", by = "Genero", grid = False)
plt.title('Diagrama de Caja y Bigotes por G칠nero')
plt.xlabel('G칠nero')
plt.ylabel('Precio')
plt.xticks(rotation = 75, ha = "right")
plt.show()
```

```python
n = df.shape[0]
bins = round(1 + 3.322 * np.log(n))
plt.hist(df["Precio"], bins = bins, color = "#7961B4", edgecolor = "black")
plt.show()
```

```python
mas_caro = df["Titulo"].loc[df["Precio"] == max(df["Precio"])]
print(f"Libro mas caro: {mas_caro.values[0]} - ${max(df['Precio'])}")


mas_barato = df["Titulo"].loc[df["Precio"] == min(df["Precio"])]
print(f"Libro mas barato: {mas_barato.values[0]} - ${min(df['Precio'])}")
```
