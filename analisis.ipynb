```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data.csv")
df.tail(5)
```

```python
df["Precio"] = df["Precio"].apply(lambda x: float(x[1:].replace(',','')))
```

```python
df["Encuadernacion"].unique()
```

```python
plt.figure(figsize=(50, 30))
bp = df.boxplot(column = "Precio", by = "Genero", grid = False)
plt.title('Diagrama de Caja y Bigotes por G칠nero')
plt.xlabel('G칠nero')
plt.ylabel('Precio')
plt.xticks(rotation = 45, ha = "right")
plt.show()
```
