```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("DataLibros.csv")
df.tail(7)
```

```python
df["Puntuaciones"].unique()
```

```python
df["Puntuaciones"].value_counts()
```

```python
colors = ['#8B75BB', '#463175', '#8958FF', '#6E3EDF','#C7BEDC']

plt.pie(df["Puntuaciones"].value_counts(), labels= df["Puntuaciones"].unique(), colors = colors,autopct='%1.1f%%')
plt.show()
```

```python
n = df.shape[0]
bins = round(1 + 3.322 * np.log(n))
plt.hist(df["Precios"], bins = bins, color = "#7961B4", edgecolor = "black")
plt.show()
```

```python
plt.scatter(df["Puntuaciones"],df["Precios"], color = "#E0B523", marker= "*")
```
